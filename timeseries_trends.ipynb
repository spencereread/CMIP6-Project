This Jupyter Notebook demonstrates the full data workflow: it loads CMIP6 datasets, extracts climate variables (temperature and precipitation) for selected geographic locations, and transforms them into a structured pandas DataFrame. It then generates time series visualizations with trend lines to highlight changes over time under different climate scenarios.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xarray as xr
from scipy import stats
from matplotlib.gridspec import GridSpec
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
from scipy.ndimage import gaussian_filter
import pandas as pd
import logging
from pathlib import Path

# Configure logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

# Define dataset directory
DATA_DIR = Path("datasets")

# Define locations for each model
LOCATIONS = {
    "UKESM1-0-LL": {"city": "London", "lat": 51.5, "lon": -0.1},
    "MIROC6": {"city": "Tokyo", "lat": 35.7, "lon": 139.7},
    "CESM2-WACCM": {"city": "Los Angeles", "lat": 34.05, "lon": -118.25},
}

# Define the experiments (ssp245 and ssp585) and variables (tas, pr)
EXPERIMENTS = ["ssp245", "ssp585","historical"]
VARIABLES = ["tas", "pr"]

def load_cmip6_data(model, experiment, variable):
    """Loads time-series data for a given model and experiment into a Pandas DataFrame with time as index."""
    file_path = DATA_DIR / f"cmip6_{variable}_{experiment}_{model}.nc"
    
    if not file_path.exists():
        logging.warning(f"File not found: {file_path}")
        return None

    logging.info(f"Loading {file_path} for {LOCATIONS[model]['city']}...")

    # Load dataset and ensure time is decoded correctly
    # Load dataset and decode times
    time_coder = xr.coders.CFDatetimeCoder(use_cftime=True)
    ds = xr.open_dataset(file_path,decode_times=time_coder)

    ds = ds.sel(time=ds.time.dt.year <= 2100)


    # Extract location details
    target_lat = LOCATIONS[model]["lat"]
    target_lon = LOCATIONS[model]["lon"]

    # Find the nearest grid point
    ds_nearest = ds.sel(lat=target_lat, lon=target_lon, method="nearest").squeeze()

    # Convert to Pandas DataFrame
    df = ds_nearest.to_dataframe().reset_index()

    # Convert time column properly
    df["time"] = pd.to_datetime(df["time"], format="%Y-%m-%d %H:%M:%S")
    df.set_index("time", inplace=True)  # Set time as index

    # Convert units
    if variable == "tas":
        df["tas"] -= 273.15  # Convert K to °C
    elif variable == "pr":
        df["pr_monthly"] = df["pr"] * 86400 * df.index.days_in_month  # Convert kg/m²/s → mm/month
        df["pr_yearly"] = df["pr"] * 86400 * 365  # Convert kg/m²/s → mm/year

    logging.info(f"Loaded dataset for {LOCATIONS[model]['city']} with {len(df)} records.")
    return df

# Load data for London (UKESM1-0-LL) and Tokyo (MIROC6)
df_london_tas_ssp245 = load_cmip6_data("UKESM1-0-LL", "ssp245", "tas")
df_london_tas_ssp585 = load_cmip6_data("UKESM1-0-LL", "ssp585", "tas")
df_london_tas_hist = load_cmip6_data("UKESM1-0-LL", "historical", "tas")
df_london_pr_ssp245 = load_cmip6_data("UKESM1-0-LL", "ssp245", "pr")
df_london_pr_ssp585 = load_cmip6_data("UKESM1-0-LL", "ssp585", "pr")
df_london_pr_hist = load_cmip6_data("UKESM1-0-LL", "historical", "pr")
#Load data for Los Angeles (CESM2-WACCM)
df_la_tas_ssp245 = load_cmip6_data("CESM2-WACCM", "ssp245", "tas")
df_la_tas_ssp585 = load_cmip6_data("CESM2-WACCM", "ssp585", "tas")
df_la_tas_hist = load_cmip6_data("CESM2-WACCM", "historical", "tas")
df_la_pr_ssp245 = load_cmip6_data("CESM2-WACCM", "ssp245", "pr")
df_la_pr_ssp585 = load_cmip6_data("CESM2-WACCM", "ssp585", "pr")
df_la_pr_hist = load_cmip6_data("CESM2-WACCM", "historical", "pr")
#Load data for Tokyo (MIROC6)
df_tokyo_tas_ssp245 = load_cmip6_data("MIROC6", "ssp245", "tas")
df_tokyo_tas_ssp585 = load_cmip6_data("MIROC6", "ssp585", "tas")
df_tokyo_tas_hist = load_cmip6_data("MIROC6", "historical", "tas")
df_tokyo_pr_ssp245 = load_cmip6_data("MIROC6", "ssp245", "pr")
df_tokyo_pr_ssp585 = load_cmip6_data("MIROC6", "ssp585", "pr")
df_tokyo_pr_hist = load_cmip6_data("MIROC6", "historical", "pr")
```

```python
# function to calculate trends
def calculate_trend(data):
    """Calculate linear trend and significance"""
    x = np.arange(len(data))
    mask = ~np.isnan(data)
    if sum(mask) <= 2:  # Need at least 3 points for regression
        return np.nan, np.nan, np.nan
    slope, intercept, r_value, p_value, std_err = stats.linregress(x[mask], data[mask])
    trend = slope * len(x)  # Total change over the period
    return trend, p_value, r_value**2

def plot_timeseries_with_trends(historical_data, ssp245_data, ssp585_data, 
                                var_name, location, unit, 
                                historical_period=(1850, 2014),
                                future_period=(2015, 2100),
                                rolling_window=10,
                                save_path=None,
                                auto_filename=False,
                                output_dir="figures"):
    """
    Plot time series with trends for historical and future projections
    
    Parameters:
    -----------
    historical_data : array-like
        Historical data values
    ssp245_data : array-like
        SSP2-4.5 scenario data values
    ssp585_data : array-like
        SSP5-8.5 scenario data values
    var_name : str
        Variable name for plot title and labels
    location : str
        Location name for plot title
    unit : str
        Unit of measurement
    historical_period : tuple
        Start and end years for historical data
    future_period : tuple
        Start and end years for projection data
    rolling_window : int
        Window size for rolling average
    save_path : str, optional
        Path to save the figure as PNG. If None and auto_filename is False, figure is not saved.
    auto_filename : bool, default False
        If True, automatically generate filename based on var_name and location
    output_dir : str, default "output"
        Directory to save figures when using auto_filename
    
    Returns:
    --------
    fig : matplotlib.figure.Figure
        The figure object
    ax : matplotlib.axes.Axes
        The axes object
    stats : dict
        Dictionary containing trend statistics
    """

    # Set plotting style
    plt.style.use('seaborn-v0_8-whitegrid')
    sns.set_context("talk")

    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Historical data
    years_hist = np.arange(historical_period[0], historical_period[1] + 1)
    if len(historical_data) != len(years_hist):
        years_hist = np.arange(historical_period[0], historical_period[0] + len(historical_data))
    
    # Add rolling mean for smoother visualization
    rolling_hist = pd.Series(historical_data).rolling(window=rolling_window, center=True).mean()
    
    # Plot historical data
    ax.plot(years_hist, historical_data, 'o', color='gray', alpha=0.3, label='Historical')
    ax.plot(years_hist, rolling_hist, color='black', linewidth=2, label=f'Historical ({rolling_window}-yr rolling)')
    
    # function to calculate trends
    def calculate_trend(data):
        """Calculate linear trend and significance"""
        x = np.arange(len(data))
        mask = ~np.isnan(data)
        if sum(mask) <= 2:  # Need at least 3 points for regression
            return np.nan, np.nan, np.nan
    
        slope, intercept, r_value, p_value, std_err = stats.linregress(x[mask], data[mask])
        trend = slope * len(x)  # Total change over the period
        return trend, p_value, r_value**2
    
    # Calculate and plot historical trend
    trend_hist, p_value_hist, r2_hist = calculate_trend(historical_data)
    if not np.isnan(trend_hist):
        # Create trend line
        y_trend = np.linspace(np.nanmean(historical_data[0:10]), 
                              np.nanmean(historical_data[0:10]) + trend_hist, 
                              len(years_hist))
        ax.plot(years_hist, y_trend, '--', color='darkblue', linewidth=1.5, 
                label=f'Hist. trend: {trend_hist:.2f} {unit} (p={p_value_hist:.3f})')
    
    # Future scenarios
    years_future = np.arange(future_period[0], future_period[1] + 1)
    if len(ssp245_data) != len(years_future):
        years_future = np.arange(future_period[0], future_period[0] + len(ssp245_data))
    
    # Add rolling means
    rolling_ssp245 = pd.Series(ssp245_data).rolling(window=rolling_window, center=True).mean()
    rolling_ssp585 = pd.Series(ssp585_data).rolling(window=rolling_window, center=True).mean()
    
    # Plot SSP2-4.5
    ax.plot(years_future, ssp245_data, 'o', color='cornflowerblue', alpha=0.3)
    ax.plot(years_future, rolling_ssp245, color='royalblue', linewidth=2, label=f'SSP2-4.5 ({rolling_window}-yr rolling)')
    
    # Calculate and plot SSP2-4.5 trend
    trend_ssp245, p_value_ssp245, r2_ssp245 = calculate_trend(ssp245_data)
    if not np.isnan(trend_ssp245):
        y_trend = np.linspace(np.nanmean(ssp245_data[0:10]), 
                              np.nanmean(ssp245_data[0:10]) + trend_ssp245, 
                              len(years_future))
        ax.plot(years_future, y_trend, '--', color='blue', linewidth=1.5,
                label=f'SSP2-4.5 trend: {trend_ssp245:.2f} {unit} (p={p_value_ssp245:.3f})')
    
    # Plot SSP5-8.5
    ax.plot(years_future, ssp585_data, 'o', color='lightcoral', alpha=0.3)
    ax.plot(years_future, rolling_ssp585, color='firebrick', linewidth=2, label=f'SSP5-8.5 ({rolling_window}-yr rolling)')
    
    # Calculate and plot SSP5-8.5 trend
    trend_ssp585, p_value_ssp585, r2_ssp585 = calculate_trend(ssp585_data)
    if not np.isnan(trend_ssp585):
        y_trend = np.linspace(np.nanmean(ssp585_data[0:10]), 
                              np.nanmean(ssp585_data[0:10]) + trend_ssp585, 
                              len(years_future))
        ax.plot(years_future, y_trend, '--', color='darkred', linewidth=1.5,
                label=f'SSP5-8.5 trend: {trend_ssp585:.2f} {unit} (p={p_value_ssp585:.3f})')
    
    # Add vertical line separating historical and projections
    ax.axvline(x=2015, color='k', linestyle='-', alpha=0.5)
    ax.text(2016.5, ax.get_ylim()[0] + 0.95*(ax.get_ylim()[1]-ax.get_ylim()[0]), 
            'Historical | Projections', ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
    
    # Set plot parameters
    ax.set_title(f'{var_name} trends for {location}', fontsize=16)
    ax.set_xlabel('Year', fontsize=14)
    ax.set_ylabel(f'{var_name} ({unit})', fontsize=14)
    ax.legend(loc='upper left', fontsize=10)
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Add trend summary text box
    textstr = (f'Historical trend: {trend_hist:.2f} {unit} (p={p_value_hist:.3f})\n'
               f'SSP2-4.5 trend: {trend_ssp245:.2f} {unit} (p={p_value_ssp245:.3f})\n'
               f'SSP5-8.5 trend: {trend_ssp585:.2f} {unit} (p={p_value_ssp585:.3f})')
    props = dict(boxstyle='round', facecolor='white', alpha=0.5)
    ax.text(0.02, 0.02, textstr, transform=ax.transAxes, fontsize=10,
            verticalalignment='bottom', bbox=props)
    
    plt.tight_layout()
    
    # Generate filename based on variable and location if auto_filename is True
    import os
    import datetime
    
    if auto_filename or save_path:
        # Ensure output directory exists
        if auto_filename:
            os.makedirs(output_dir, exist_ok=True)
            
            # Format var_name and location for filename (replace spaces, special chars)
            var_clean = var_name.lower().replace(' ', '_').replace('/', '_')
            loc_clean = location.lower().replace(' ', '_').replace(',', '').replace('/', '_')
            
            # Add timestamp for uniqueness
            timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
            
            # Create filename
            filename = f"{var_clean}_{loc_clean}.png"
            final_path = os.path.join(output_dir, filename)
        else:
            final_path = save_path
            save_dir = os.path.dirname(final_path)
            if save_dir and not os.path.exists(save_dir):
                os.makedirs(save_dir)
        
        # Save with high dpi for better quality
        plt.savefig(final_path, dpi=300, bbox_inches='tight', format='png')
        print(f"Figure saved to: {final_path}")
    
    return fig, ax, {
        'historical_trend': trend_hist,
        'ssp245_trend': trend_ssp245,
        'ssp585_trend': trend_ssp585,
        'historical_p': p_value_hist,
        'ssp245_p': p_value_ssp245,
        'ssp585_p': p_value_ssp585,
        'historical_r2': r2_hist,
        'ssp245_r2': r2_ssp245,
        'ssp585_r2': r2_ssp585
    }
```

```python
def compute_yearly_average(df, variable):
    df_var = df[[variable]].copy()
    df_yearly = df_var.resample("YE").mean()
    df_std = df_var.resample("YE").std()*0.25
    return df_yearly[variable], df_std[variable]
```

```python
anom_ssp245_tas, std_ssp245_tas = compute_yearly_average(df_london_tas_ssp245,"tas")
anom_ssp585_tas, std_ssp585_tas = compute_yearly_average(df_london_tas_ssp585,"tas")
anom_hist_tas, std_hist_tas = compute_yearly_average(df_london_tas_hist,"tas")
anom_ssp245_pr, std_ssp245_pr = compute_yearly_average(df_london_pr_ssp245,"pr_monthly")
anom_ssp585_pr, std_ssp585_pr = compute_yearly_average(df_london_pr_ssp585,"pr_monthly")
anom_hist_pr, std_hist_pr = compute_yearly_average(df_london_pr_hist,"pr_monthly")
```

```python
plot_timeseries_with_trends(anom_hist_tas, anom_ssp245_tas,anom_ssp585_tas,"Air Temperature","London","°C",auto_filename=True)
plot_timeseries_with_trends(anom_hist_pr, anom_ssp245_pr, anom_ssp585_pr, "Precipitation", "London", "mm/month",auto_filename=True)
```
